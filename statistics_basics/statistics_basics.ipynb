# Задание 1.

```python
horses_df = pd.read_csv('horse_data.csv', usecols=[0, 1, 3, 4, 5, 6, 10, 22],
                    names=['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 
                    'temperature of extremities', 'pain', 'outcome'], na_values='?')
horses_df['Age'] = horses_df['Age'].replace(9, 2)
horses_df
```

# Задание 2.

```python
# Мера центральной тенденции.
print('Среднее арифметическое:')
print(horses_df.mean().round(2))
print('')
print('Медиана:')
print(horses_df.median())
print('')
print('Мода:')
print(horses_df.mode().transpose()[0])
```

### Выводы:
1. Большинство лечений сопровождалось оперативным вмешательством.
1. В большей степени заболевали взрослые особи.
1. Заболевания сопровождались в большинстве случаев небольшим повышением ректальной температуры и температуры конечностей.
1. Отмечается повышение пульса, что в совокупности с другими фактами может свидетельствовать о шоке или о болевом синдроме.
1. Частота дыхания у большинства особей повышена, однако данный факт лучше сочетать с другими более значительными, поскольку в данном случае имеется большая склонность к флуктуациям.
1. Большая часть обследуемых лошадей имела лёгкие прерывыстые болевые ощущения.
1. Большая часть лошадей выжила \\(◕‿◕)/

```python
# Мера разрбоса.
print('Стандартное отклонение:')
print(horses_df.std().round(2))
print('Дисперсия:')
print((horses_df.std()**2).round(2))
```

Менее всего, очевидно, разбросаны значения, которые представлены в виде общей оценки состояния, полученной из первичных данных или из какого-либо факта. Также в целом неудивительно, что далеко разбросаны значения пульса и частоты дыхания, поскольку обе данные характеристики очень неустойчивы даже с учётом полностью здоровой особи. Разбросы температур, ректальной и конечностей, демонстрируют относительную устойчивость этой характеристики в рамках данных измерений у разных особей.

```python
# Выбросы.
q1 = horses_df.quantile(0.25)
q3 = horses_df.quantile(0.75)

# Межквартильный размах.
iqr = q3 - q1

# Границы.
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# Определяем выбросы.
# Не учитываем столбцы, в которых указаны оценочные вторичные данные.
determine_outliers = horses_df.loc[:, ['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities']]\
                        [horses_df.apply(lambda row: (row < lower_bound) | (row > upper_bound), axis=1)].dropna(thresh=1)
determine_outliers
```

# Задание 3.
### При выполнении данного задания подбор значений для заполнения пустых ячеек в датафрейме выполнялся по мере центральной тенденции искомого датафрейма, то есть чем ближе величина среднего арифметического, медианы и моды заполненного столбца к тем же величинам искомого столбца, тем лучше.

```python
# Создаём датафрейм для заполнения.
import copy
horses_df_filled = copy.deepcopy(horses_df)
```

```python
# Количество пропусков в столбцах.
horses_df_filled.isnull().sum()
```

Столбцы 'surgery?' и 'outcome' заполняем модой, поскольку в них отсутсвует всего лишь одно значение и сами столбцы состоят из вторичных оценок первичных данных.

```python
# Столбцы 'surgery?' и 'outcome'.
horses_df_filled['surgery?'] = horses_df_filled['surgery?'].fillna(horses_df_filled['surgery?'].mode()[0])
horses_df_filled['outcome'] = horses_df_filled['outcome'].fillna(horses_df_filled['outcome'].mode()[0])

# # Проверка.
# print(horses_df_filled['surgery?'].isnull().sum())
# print(horses_df_filled['outcome'].isnull().sum())
```

Столбец 'pain' заполняем по группировке столбцов наличия операции и исхода лечения. В данном случае наличие боли может быть причиной оперативного вмешательства, что в свою очередь в совокупности с болью может привести к тому или иному исходу.

```python
# Столбец 'pain'.
horses_df_filled['pain'] = horses_df_filled['pain'].fillna(horses_df_filled.groupby(['surgery?', 'outcome'])['pain'].transform('median'))

# # Проверка.
# print(horses_df_filled['pain'].isnull().sum())
# print(horses_df_filled['pain'].mean())
# print(horses_df_filled['pain'].median())
# print(horses_df_filled['pain'].mode()[0])
```

Столбец 'rectal temperature' заполняем группировкой датафрейма по болевому синдрому и наличию оперативного вмешательства. Из наличия боли следует выполнение оперативного лечения, что в свою очередь связано с повышением температуры.

```python
# Подбор.
rec_temp_ds = horses_df_filled['rectal temperature'].fillna(horses_df_filled.groupby(['pain', 'surgery?'])['rectal temperature'].transform('median'))
print(rec_temp_ds.mean())
print(rec_temp_ds.median())
print(rec_temp_ds.mode()[0])
```

```python
# Столбец 'rectal temperature'.
horses_df_filled['rectal temperature'] = horses_df_filled['rectal temperature'].fillna(horses_df_filled.groupby(['pain', 'surgery?'])['rectal temperature'].transform('median'))

# # Проверка.
# print(horses_df_filled['rectal temperature'].isnull().sum())
```

Столбец 'temperature of extremities' заполняем группировкой датафрейма по болевому синдрому и ректальной температуре. Взаимосвязь температур очевидна, а болевой синдром может сопровождаться изменением температуры.

```python
# Подбор.
temp_extr_ds = horses_df['temperature of extremities'].fillna(horses_df.groupby(['pain', 'rectal temperature'])['temperature of extremities'].transform('median'))
print(temp_extr_ds.mean())
print(temp_extr_ds.median())
print(temp_extr_ds.mode()[0])
```

```python
# Столбец 'temperature of extremities'.
horses_df_filled['temperature of extremities'] = horses_df_filled['temperature of extremities'].fillna(horses_df_filled.groupby(['pain'])['temperature of extremities'].transform('median'))

# # Проверка.
# print(horses_df_filled['temperature of extremities'].isnull().sum())
```

Столбец 'pulse' заполняем по болевым ощущениям, возрасту и наличию оперативного вмешательства. Пульс чувствителен к возрасту и боли, а также наличие операции может свидетельствовать о тяжелом состоянии лошади, что тоже сказывается на пульсе.

```python
pulse_ds = horses_df['pulse'].fillna(horses_df.groupby(['Age', 'surgery?', 'pain'])['pulse'].transform('median'))
print(pulse_ds.mean())
print(pulse_ds.median())
print(pulse_ds.mode()[0])
```

```python
# Столбец 'pulse'.
horses_df_filled['pulse'] = horses_df_filled['pulse'].fillna(horses_df_filled.groupby(['Age', 'surgery?', 'pain'])['pulse'].transform('median'))

# # Проверка.
# print(horses_df_filled['pulse'].isnull().sum())
```

Столбец 'respiratory rate' заполняем сначала по группировке по пульсу, оставшиеся значения дополняем медианой искомого столбца.

```python
resp_rate_ds = horses_df['respiratory rate'].fillna(horses_df.groupby(['pulse'])['respiratory rate'].transform('median')).fillna(horses_df['respiratory rate'].median())
print(resp_rate_ds.isnull().sum())
print(resp_rate_ds.mean())
print(resp_rate_ds.median())
print(resp_rate_ds.mode()[0])
```

```python
# Столбец 'respiratory rate'.
horses_df_filled['respiratory rate'] = horses_df_filled['respiratory rate'].fillna(horses_df_filled.groupby(['pulse'])['respiratory rate'].transform('median')).fillna(horses_df_filled['respiratory rate'].median())

# # # Проверка.
# print(horses_df_filled['respiratory rate'].isnull().sum())
```
