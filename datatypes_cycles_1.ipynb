```python
def find_mid_letters(word: str) -> str:
    """ Returns two middle letters of the given word
    if word is even, or one middle letter if it's odd. """
    
    word_len = len(word)
    if word_len == 0:
        return 'Вы ввели пустую строку.'
    
    even_ind = (word_len % 2 == 0)
    mid = word_len // 2
    answer = ''
    if even_ind:
        answer = word[mid-1] + word[mid]
    else:
        answer = word[mid]
    return answer
    
# Test.
assert find_mid_letters('abrakadabra') == 'a'
assert find_mid_letters('python') == 'th'
assert find_mid_letters('') == 'Вы ввели пустую строку.'
assert find_mid_letters('test') == 'es'
assert find_mid_letters('testing') == 't'
```

# Задание 3

```python
def acquaint(boys: list, girls: list) -> str:
    """ Makes couples using boys and girls lists. """
    
    couples = []
    len_boys = len(boys)
    len_girls = len(girls)
    s_boys = sorted(boys)
    s_girls = sorted(girls)
    if len_boys == 0 and len_girls == 0:
        print("Оба списка пустые!")
    
    if len(boys) != len(girls):
        return "Внимание, кто-то может остаться без пары. "
    else:
        couples = list(zip(s_boys, s_girls))
    
    str_couples = prepare_to_print(couples)
    
    return str_couples


def prepare_to_print(couples: list) -> str:
    """ Prepares couples list to be properly printed. """

    result = 'Идеальные пары:'
    for couple in couples:
        result += f'\n{couple[0]} и {couple[1]}'
    
    return result


# Test 1.
boys_1 = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls_1 = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
str_couples = 'Идеальные пары:\nAlex и Emma\nArthur и Kate\nJohn и Kira\nPeter и Liza\nRichard и Trisha'
assert acquaint(boys_1, girls_1) == str_couples

# Test 2.
boys_2 = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
girls_2 = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
assert acquaint(boys_2, girls_2) == "Внимание, кто-то может остаться без пары. "
```

```python
def find_appr_cels_temperature(countries_temperature: list) -> str:
    """ Defines average temperature in Celsius for countries. """
    
    if len(countries_temperature) == 0:
        return "Программа не может обработать пустой список."
        
    aver_temps = []
    for elem in countries_temperature:
        aver_temps_elem = []
        
        # Appending country.
        aver_temps_elem.append(elem[0])
        
        # Calculating average temperature.
        aver_temp_far = find_aver(elem[1])
        
        # Converting to Celsius.
        aver_temp_cels = convert_to_celsius(aver_temp_far)
        
        # Appending temperature.
        aver_temps_elem.append(aver_temp_cels)
        
        # Appending country-average temperature list in common list.
        aver_temps.append(aver_temps_elem)
    
    # Output to print.
    result = prepare_to_print(aver_temps)
    return result
 
    
def find_aver(list_of_temps: list) -> int:
    """ Calculates average temperature. """
    
    list_len = len(list_of_temps)
    aver_temp_far = sum(list_of_temps) / list_len
    return aver_temp_far
    
    
def convert_to_celsius(far_temp: int) -> int:
    """ Coverts Farenheit degrees to Celsuis degrees."""
    
    cels_temp = (5 / 9) * (far_temp - 32)
    return cels_temp


def prepare_to_print(appr_temps: list) -> str:
    """ Prepares couples list to be properly printed. """

    result = 'Средняя температура в странах:'
    for elem in appr_temps:
        result += f'\n{elem[0]}  -  {elem[1]:.1f} С'
    
    return result


# Test 1.
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

str_temp = "Средняя температура в странах:\nThailand  -  23.9 С\nGermany  -  13.8 С\nRussia  -  3.7 С\nPoland  -  12.0 С"
assert find_appr_cels_temperature(countries_temperature) == str_temp

# Test 2.
countries_temperature = []
assert find_appr_cels_temperature(countries_temperature) == "Программа не может обработать пустой список."
```

```python
def find_sum_of_views(logs_lst: list) -> int:
    """ Calculates views quantity. """

    count = 0
    for log in logs_lst:
        count += int(log[2])

    return count


def find_uniq_users(logs_lst: list) -> list:
    """ Finds unic users in logs. """

    users = []
    for elem in logs_lst:
        if elem[1] not in users:
            users.append(elem[1])

    return users


def get_sublists_list(views_log: list) -> list:
    """ Splits strings into sublists and makes common list. """

    logs_lst = []
    for log in views_log:
        date_user_views = log.split(',')
        logs_lst.append(date_user_views)

    return logs_lst


def find_aver_views_qty(views_log: list) -> str:
    """ Finds average quantity of views made by unique viewers. """
    
    logs_lst = get_sublists_list(views_log)
    users = find_uniq_users(logs_lst)
    views_sum = find_sum_of_views(logs_lst)
    
    aver_views = views_sum / len(users)
    return f"Среднее количество просмотров на уникального пользователя: {aver_views:.2f}"


# Test 1.
stream_1 = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

str_1 = "Среднее количество просмотров на уникального пользователя: 23.25"
assert find_aver_views_qty(stream_1) == str_1

# Test 2.
stream_2 = [
    '2018-01-01,user100,150',
    '2018-01-07,user99,205',
    '2018-03-29,user1001,81'
]
str_2 = "Среднее количество просмотров на уникального пользователя: 145.33"
assert find_aver_views_qty(stream_2) == str_2
```

```python
def get_uniq_nums(sep_num_lst: list) -> list:
    """ Returns a list of unique numbers. """
    
    uniq_nums = []
    for num in sep_num_lst:
        if num in uniq_nums:
            continue
        else:
            uniq_nums.append(num)
    
    uniq_nums.sort()        
    return uniq_nums


def get_nums_freq(uniq_nums: list, sep_num_lst: list) -> list:
    """ Returns list of numbers that repeat in the given list. """
    
    nums_freq_lst = []
    for num in uniq_nums:
        
        count = 0
        for item in sep_num_lst:
            if item == num:
                count += 1

        nums_freq_lst.append([num, count])
    
    return nums_freq_lst
    
    
def get_list_of_repeated_nums(sep_num_lst: list) -> list:
    """ Returns sorted list of nums that repeated in the given string. """
    
    sep_nums_int_lst = list(map(int, sep_num_lst))
    uniq_nums = get_uniq_nums(sep_nums_int_lst)
    nums_freq_lst = get_nums_freq(uniq_nums, sep_nums_int_lst)
    
    result = []
    for elem in nums_freq_lst:
        if elem[1] == 1:
            continue
        else:
            result.append(elem[0])
            
    result_str = ' '.join(list(map(str, result)))
    return result_str
    

# Test. ----------------------------------------------------------------------------------------------    
inp_test_1 = '4 8 0 3 4 2 0 3'.split(' ')
assert get_list_of_repeated_nums(inp_test_1) == '0 3 4'
inp_test_2 = '10 15 15 103 200 200 200 1 1 1 1 1 2 2 2'.split(' ')
assert get_list_of_repeated_nums(inp_test_2) == '1 2 15 200'
# ----------------------------------------------------------------------------------------------------


# Interface.
flag = True
while flag:
    print("Программа получает на вход последовательность чисел.")
    print("Для начала работы введите 's'.")
    print("Для выхода из программы введите 'q'.")
    
    inp = input("Ваш ввод: ")
    if inp == 'q':
        break
    elif inp == 's':
        
        while True:

            print("Введите последовательность чисел по одному.")
            print("Когда закончите ввод, введите 'f' и программа выведет отсортированный список, " +
                "состоящий из элементов, которые повторялись в введённой последовательности.")
            nums = []
            
            while True:
                
                input_num = input("Введите число: ")
                
                if input_num == 'f':
                    result = get_list_of_repeated_nums(nums)
                    print(result)
                    break
                elif len(input_num.split()) > 1:
                    print("Неверный ввод.")
                    continue
                elif len(input_num) == 0:
                    print("Вы ввели пустую строку.")
                    continue
                else:
                    nums.append(int(input_num))
                
            print("Если хотите продолжить введите любой символ (кроме 'q').")
            print("В противном случае введите 'q'.")
            inner_inp = input("Ваш ввод: ")
            if inner_inp == 'q':
                flag = False
                break

    else:
        print("Такой команды нет.")
```
