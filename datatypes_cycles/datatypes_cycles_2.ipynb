```python
# Main func.
def get_uniq_label(ids):
    """ Returns unique geo-labels of users. """
    
    labels_list = [label for key in ids.keys() for label in ids[key]]
    
    return set(labels_list)
    

# Test.
ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}
assert get_uniq_label(ids) == {98, 35, 15, 213, 54, 119}
```

# Задание 2

```python
def get_queries_len_dict(queries):
    """
    Returns dictionary with quantity of the words in queries as key and
    percent of each type in all queries as values. 
    """
    
    # Ratio for converting values to percents.
    ratio = 100 / len(queries)
    
    queries_len_lst = [len(query.split()) for query in queries]
    queries_len_set = sorted(list((set(queries_len_lst))))
    
    # Not the clearest choice but it's to learn dict-comprehensions :D.
    queries_len_dict = {
            uniq_len: (ratio * queries_len_lst.count(uniq_len)) for uniq_len in queries_len_set for query in queries_len_lst
        }
    return queries_len_dict
    
    
    
def prepare_to_print(queries_len_dict):
    """ Prepares data to be printed properly. """
    
    result = ''
    for qty, percent in queries_len_dict.items():
        result += f'Поисковых запросов, содержащих {qty} слов(а): {percent:.2f}%\n'
    
    return result


# Main func.
def get_words_distr_percent(queries):
    """ Determines the distribution of the words in queries. """
    
    if len(queries) == 0:
        return 'Пустой список.'
    
    queries_len_dict = get_queries_len_dict(queries)
    result = prepare_to_print(queries_len_dict)
    
    return result


# Test 1.
queries_1 = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
right_str = 'Поисковых запросов, содержащих 2 слов(а): 42.86%\nПоисковых запросов, содержащих 3 слов(а): 57.14%\n'
assert get_words_distr_percent(queries_1) == right_str

# Test 2.
queries_2 = []
assert get_words_distr_percent(queries_2) == 'Пустой список.'
```

# Задание 3

```python
# Main func.
def get_companies_roi(results):
    """ Returns companies cost and revenue dictionary supplemented with ROI. """
    
    if not results:
        return 'Пустой словарь.'

    revenue = 0
    cost = 0
    for company in results.values():
        revenue = company['revenue']
        cost = company['cost']
        roi = 100 * (revenue / cost - 1)
        
        # Adding ROI.
        company['ROI'] = round(roi, 2)
        
    return results


# Test 1.
results_1 = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}
right_result = {
    'adwords': {'cost': 34, 'revenue': 35, 'ROI': 2.94},
    'facebook': {'cost': 110, 'revenue': 103, 'ROI': -6.36},
    'twitter': {'cost': 24, 'revenue': 11, 'ROI': -54.17},
    'vk': {'cost': 98, 'revenue': 103, 'ROI': 5.1},
    'yandex': {'cost': 153, 'revenue': 179, 'ROI': 16.99}
    }
assert get_companies_roi(results_1) == right_result

# Test 2.
results_2 = {}
assert get_companies_roi(results_2) == 'Пустой словарь.'
```

```python
def determine_max_sales_channel(stats):
    """ Returns name of the channel that has the maximal sales value. """
    
    if not stats:
        return 'Пустой словарь.'
    
    tuple_stats = []
    for channel, value in stats.items():
        tuple_stats.append((channel, value))
    
    tuple_stats.sort(key=lambda elem: elem[1], reverse=True)
    result = []
    for elem in tuple_stats:
        if result and (elem[1] < stats[result[0]]):
                break
        result.append(elem[0])
        
    result_str = ', '.join(result)
    if len(result) > 1:
        # If sales value equal for some of channels.
        return f'Максимальный объём продаж на рекламных каналах: {result_str}'
    return f'Максимальный объём продаж на рекламном канале: {result_str}'
        
    
# Test 1.
stats_1 = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}
assert determine_max_sales_channel(stats_1) == 'Максимальный объём продаж на рекламном канале: vk'

# Test 2.
stats_2 = {}
assert determine_max_sales_channel(stats_2) == 'Пустой словарь.'

# Test 3.
stats_3 = {'facebook': 55}
assert determine_max_sales_channel(stats_3) == 'Максимальный объём продаж на рекламном канале: facebook'

# Test 4.
stats_4 = {'facebook': 120, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}
assert determine_max_sales_channel(stats_4) == 'Максимальный объём продаж на рекламных каналах: facebook, vk'
```

```python
# Main func.
def get_leveled_dict(some_lst):
    """ Returns dictionary that leveled in the extent that equals the length of the given list. """
    
    lst_len = len(some_lst)
    some_dict = dict()

    # Crutch №1.
    if lst_len == 1:
        return {some_lst[0]}

    # Main code.
    root = some_dict
    for i in range(lst_len-1):

        # Crutch №2.
        if i == 0:
            some_dict[some_lst[i]] = some_lst[i+1]
            continue

        root[some_lst[i-1]] = {some_lst[i]: some_lst[i+1]}
        root = root[some_lst[i-1]]
    
    return some_dict


# Test 1.
my_list_1 = ['2018-01-01', 'yandex', 'cpc', 100]
assert  get_leveled_dict(my_list_1) == {'2018-01-01': {'yandex': {'cpc': 100}}}

# Test 2.
my_list_2 = ['a', 'b', 'c', 'd', 'e', 'f']
assert  get_leveled_dict(my_list_2) == {'a': {'b': {'c': {'d': {'e': 'f'}}}}}

# Test 3.
my_list_3 = []
assert  get_leveled_dict(my_list_3) == {}

# Test 4.
my_list_4 = ['a']
assert  get_leveled_dict(my_list_4) == {'a'}

# Test 5.
my_list_5 = ['a', 'b', 'c', 'd', 'e']
assert  get_leveled_dict(my_list_5) == {'a': {'b': {'c': {'d': 'e'}}}}
```

```python
# Data ---------------------------------------------------------------------
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},   
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},    
    ]
}

# --------------------------------------------------------------------------


def prepare_to_print(result_lst: list) -> str:
    """ Prepares data to be printed properly. """
    
    result = []
    for elem in result_lst:
        qty = 3 * elem[1]
        string = f'{elem[0].capitalize()}: {qty} {elem[2]}'
        result.append(string)

    return '\n'.join(result)


# Main func.
def make_list_of_food(qty: int) -> str:
    """ Returns list of food that needed to cook the dishes in depending on quantity of portions. """
    
    ingr_qty_meas = []
    for dish in cook_book:
        ingr_qty_meas += cook_book[dish]
    
    result_lst = []
    lst_len = len(ingr_qty_meas)
    for i in range(lst_len):
        acc = ingr_qty_meas[i]['quantity']
        ingr = ingr_qty_meas[i]['ingridient_name']
        meas = ingr_qty_meas[i]['measure']
        
        for j in range(i, lst_len):
            ingr_match = (ingr == ingr_qty_meas[j]['ingridient_name'])
            meas_match = (meas == ingr_qty_meas[j]['measure'])
            
            if ingr_match and meas_match:
                acc += ingr_qty_meas[j]['quantity']
                
        result_lst.append([ingr, acc, meas])
    
    result = prepare_to_print(result_lst)
    return result
                              

# Interface.
while True:
    
    print('Программа выводит количество ингридиентов необходимых для приготовления следующих блюд: ' +
        'салат, пицца, лимонад.')
    print('Для начала работы введите "s". Для выхода введите "q".')
    
    inp = input('Ваш ввод: ')
    if inp == 'q':
        print('Завершаем программу.')
        break
    elif inp == 's':
        qty = int(input('Введите количество порций: '))

        if int(qty) <= 0:
            print(f'Невозможно выполнить действие для количества {qty}. Пожалуйста, введите корректные данные.')
            continue

        print(make_list_of_food(qty))
    else:
        print('Некорректный ввод.')
```
