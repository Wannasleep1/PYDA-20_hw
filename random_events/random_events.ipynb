## Решение через среднюю величину.

```python
import numpy as np
```

```python
def spin_the_slot_machine():
    import copy
    qty = 10**10
    initial_cash = 10**6
    difference = copy.deepcopy(initial_cash)
    diviser = 1
    
    # В цикле иммитируем процесс прокрутки игрального автомата.
    for i in range(qty):
        # Проверяем остались ли деньги на счету.
        if initial_cash == 0:
            break
        diviser = i
        
        # Вычитаем из начальной суммы по 1 рублю за прокрутку.
        initial_cash -= 1
        
        # Крутим игральный автомат.
        result_of_rolling = np.random.randint(0, 1000, 1)[0]

        # Проверяем есть ли выиграшная комбинация.
        if result_of_rolling == 777:
            initial_cash += 200
        elif result_of_rolling == 999:
            initial_cash += 100
        elif result_of_rolling == 555:
            initial_cash += 50
        elif result_of_rolling == 333:
            initial_cash += 15
        elif result_of_rolling == 111:
            initial_cash += 10
        elif result_of_rolling % 10 == 7:
            if result_of_rolling % 100 == 77:
                initial_cash += 5
            else:
                initial_cash += 3
        elif result_of_rolling % 10 == 0:
            if (result_of_rolling != 0) and (result_of_rolling % 100 == 0):
                initial_cash += 2
            else:
                initial_cash += 1
    
    return [initial_cash-difference, diviser]


# result = spin_the_slot_machine()
# print(result[0] / result[1])
```

В результате проделанного моделирования прокрутки игрального автомата определённое количество раз с определённой изначальной суммой денег определено:
1. При увеличении числа повторений игрок всегда теряет все свои деньги.
1. Для каждой суммы денег существует такое число прокруток, при котором дальнейшее их увеличение не будет приводить к стабилизации средней убыли, поскольку деньги буду кончатся намного раньше достижения заданного числа прокруток.
1. Средняя убыль составляет примерно 20 коппек за прокрутку.
1. Средняя убыль определялась по следующему факту, определённому экспериментально: при росте числа повторений средняя убыль стремится к диапазону от 0 до -1 рубля за прокрутку; однако если сделать так, чтобы величины начального банка хватало на число прокруток близкого к 500000 и более, то заметим что действительно средняя убыль стремится к 20 копейкам.
1. Игра может быть выгодна с некоторой вероятностью при малом количестве прокруток.

В случае эксперимента использовались значения: число повторений 10^10, начальный банк 10^6 рублей. По результатам 10 прокруток с указанными значениями получили величины: -0.198, -0.202, -0.200, -0.205, -0.202, -0.202, -0.202, -0.205, -0.201, -0.198. Делаем вывод, что убыль cопоставима примерно 0.2 рублям, что как раз равняется 20 копейкам.
Также были проверены меньшие значения для обеих величин, что подтвердило предположение о средней убыли 20 копеек.

## Теоретическое решение.

Воспользуемся формулой среднего значения, тогда при вероятности всех величин равной 0.001 имеем:

Хср = 0.001 * (200 + 100 + 50 + 15 + 10 + 9 * 5 + 90 * 1 + 90 * 3 + 10 * 2 - 1000) = - 0.2

То есть выносим за скобку вероятность выпадения любой величины (равномерное распределение), далее складываем все возможные суммы, которые могут нам выпасть, и вычитаем 1000 рублей, которые мы потратим на прокрутки. По итогу получаем тот же убыток в 20 копеек, который мы получили при моделировании.
