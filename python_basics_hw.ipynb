# Задание 1

```python
def find_longest_phrase(a: str, b: str) -> str:
    """ Finds the longest string of two ones. """
    
    longest = ''
    not_longest = ''
    a_len = len(a)
    b_len = len(b)
    if a_len > b_len:
        longest = '1'
        not_longest = '2'
    elif a_len < b_len:
        longest = '2'
        not_longest = '1'
    else:
        return "Фразы равной длины."
    
    return f"Фраза {longest} длиннее фразы {not_longest}."

# Test 1.
a = 'Насколько проще было бы писать программы, если бы не заказчики'
b = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
assert find_longest_phrase(a, b) == "Фраза 1 длиннее фразы 2."

# Test 2.
b = 'Насколько проще было бы писать программы, если бы не заказчики'
a = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
assert find_longest_phrase(a, b) == "Фраза 2 длиннее фразы 1."

# Test 3.
a = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
b = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
assert find_longest_phrase(a, b) == "Фразы равной длины."
```

# Задание 2

```python
def is_leap_year(year: int) -> str:
    """ Checks if the given year is leap or not. """
    
    modulo = year % 4
    if modulo != 0:
        return "Обычный год."
    return "Високосный год."

# Test 1.
year = 2020
assert is_leap_year(year) == "Високосный год."

# Test 2.
year = 2019
assert is_leap_year(year) == "Обычный год."
```

```python
months = [
    'январь', 'февраль', 'март', 'апрель', 'май', 
    'июнь', 'июль', 'август', 'сентябрь', 'октябрь',
    'ноябрь', 'декабрь',
    ]


def is_correct_data(day: int, month: str) -> bool:
    """ Checks if data is correct. """
    
    flag = False
    day_crts = (1 <= day <= 31)
    if (not day_crts) or (month not in months):
        flag = True
    return flag
    
    
def define_zodiac_sign(day: str, month: str) -> str:
    """ Defines a zodiac sign by the day and the month. """
    
    zodiac_sign = ''
    day = int(day)
    month = month.lower()
    
    # Input data correctness check_1.
    data_crts = is_correct_data(day, month)
    if data_crts:
        return 'Введены некорректные данные. Попробуйте снова.'
    
    # Defining zodiac sign.
    if month == 'март':
        if day >= 21:
            zodiac_sign = 'овен'
        else:
            zodiac_sign = 'рыбы'
    elif month == 'апрель':
        if day == 31:
            data_crts = True
        elif day >= 21:
            zodiac_sign = 'телец'
        else:
            zodiac_sign = 'овен'
    elif month == 'май':
        if day >= 22:
            zodiac_sign = 'близнецы'
        else:
            zodiac_sign = 'телец'
    elif month == 'июнь':
        if day == 31:
            data_crts = True
        elif day >= 22:
            zodiac_sign = 'рак'
        else:
            zodiac_sign = 'близнецы'
    elif month == 'июль':
        if day >= 24:
            zodiac_sign = 'лев'
        else:
            zodiac_sign = 'рак'
    elif month == 'август':
        if day >= 24:
            zodiac_sign = 'дева'
        else:
            zodiac_sign = 'лев'
    elif month == 'сентябрь':
        if day == 31:
            data_crts = True
        elif day >= 24:
            zodiac_sign = 'весы'
        else:
            zodiac_sign = 'дева'
    elif month == 'октябрь':
        if day >= 24:
            zodiac_sign = 'скорпион'
        else:
            zodiac_sign = 'дева'
    elif month == 'ноябрь':
        if day == 31:
            data_crts = True
        elif day >= 23:
            zodiac_sign = 'стрелец'
        else:
            zodiac_sign = 'скорпион'
    elif month == 'декабрь':
        if day >= 22:
            zodiac_sign = 'козерог'
        else:
            zodiac_sign = 'стрелец'
    elif month == 'январь':
        if day >= 21:
            zodiac_sign = 'водолей'
        else:
            zodiac_sign = 'козерог'
    elif month == 'февраль':
        if day > 29:
            data_crts = True
        elif day >= 20:
            zodiac_sign = 'рыбы'
        else:
            zodiac_sign = 'водолей'
    
    
    # Input data correctness check_2.
    if data_crts:
        return 'Введены некорректные данные. Попробуйте снова.'
    
    return f'Ваш знак зодиака: {zodiac_sign.capitalize()}.'


while True:
    print('Для того, чтобы узнать свой знак зодиака, необходимо будет указать ' +\
    'месяц и день своего рождения. (Введите символ "q", если хотите завершить.)')
    month = input('Введите месяц своего рождения: ')
    day = input('Введите день своего рождения: ')
    if month == 'q' or day == 'q':
        break
    else:
        answer = define_zodiac_sign(day, month)
        print(answer)

# Test 1.
assert define_zodiac_sign('-3', 'небритябрь') == 'Введены некорректные данные. Попробуйте снова.'
assert define_zodiac_sign('1000', 'небритябрь') == 'Введены некорректные данные. Попробуйте снова.'
assert define_zodiac_sign('0', 'небритябрь') == 'Введены некорректные данные. Попробуйте снова.'
assert define_zodiac_sign('22', '') == 'Введены некорректные данные. Попробуйте снова.'

# Test 2.
assert define_zodiac_sign('29', 'октябрь') == 'Ваш знак зодиака: Скорпион.'
assert define_zodiac_sign('24', 'ноябрь') == 'Ваш знак зодиака: Стрелец.'
assert define_zodiac_sign('29', 'август') == 'Ваш знак зодиака: Дева.'
```

# Задание 4

```python
def choose_box_size(a: int, b: int, c: int) -> str:
    """ Chooses box size using product sizes. """
    
    answer = ''
    list_of_sizes = [a, b, c]
    for size in list_of_sizes:
        if size <= 0:
            return 'Размеры не могут быть равны 0 и меньше.'
            
    crit_1 = [elem < 15 for elem in list_of_sizes]
    crit_2 = [50 > elem >= 15 for elem in list_of_sizes]
    crit_3 = [200 > elem >= 50 for elem in list_of_sizes]
    crit_4 = [elem >= 200 for elem in list_of_sizes]
    
    if all(crit_1):
        answer = "Коробка №1"
    elif any(crit_4):
        answer = "Упаковка для лыж"
    elif any(crit_3):
        answer = "Стандартная коробка №3"
    else:
        answer = "Коробка №2"
        
    return answer

# Test.
assert choose_box_size(15, 50, 15) == "Стандартная коробка №3"
assert choose_box_size(65, 75, 55) == "Стандартная коробка №3"
assert choose_box_size(200, 75, 55) == "Упаковка для лыж"
assert choose_box_size(49, 13, 40) == "Коробка №2"
assert choose_box_size(14, 13, 12) == "Коробка №1"
assert choose_box_size(0, -3, 12) == 'Размеры не могут быть равны 0 и меньше.'
```

# Задание 5

```python
def is_lucky(number: str) -> str:
    """ Defines if number of ticket is lucky or not. """
    num_len = len(number)
    if num_len == 0 or num_len % 2 != 0:
        return 'Введены неверные данные. Длина номера не может быть нечетной или равно 0.'
    
    mid = len(number) // 2
    list_of_num = list(map(int, list(number)))
    first_half = sum(list_of_num[:mid])
    second_half = sum(list_of_num[mid:])
    
    result = 0
    if first_half == second_half:
        result = 'Счастливый билет!'
    else:
        result = 'Несчастливый билет...'
    
    return result

while True:
    print("Программа скажет счастливый ли у вас билет или нет.")
    print("Нельзя вводить номер, состоящий из нечетного числа цифр, или " + 
    "пустую строку.")
    print("Для выхода из программы введите символ 'q'.")
    input_num = input('Введите сюда ваш номер билета: ')
    if input_num == 'q':
        break
    else:
        result = is_lucky(input_num) 
        print(result)

# Test.
assert is_lucky('123321') ==  'Счастливый билет!'
assert is_lucky('123121') == 'Несчастливый билет...'
assert is_lucky('') == 'Введены неверные данные. Длина номера не может быть нечетной или равно 0.'
assert is_lucky('123') == 'Введены неверные данные. Длина номера не может быть нечетной или равно 0.'
```

```python
def calc_square_area() -> int:
    """ Calculates area of a square by the given side length. """
    
    side_len = float(input('Введите длину стороны квадрата в сантиметрах: '))
    area = side_len ** 2
    return area


def calc_circle_area() -> int:
    """ Calculates area of a circle by the given radius length. """
    
    radius = float(input('Введите длину радиуса круга в сантиметрах: '))
    pi = 3.1416
    area = pi * (radius ** 2)
    return area
    
    
def calc_triangle_area() -> int:
    """ Calculates area of trinagle by the Geron Method. """
    
    a = float(input('Введите длину первой стороны в сантиметрах: '))
    b = float(input('Введите длину второй стороны в сантиметрах: '))
    c = float(input('Введите длину третьей стороны в сантиметрах: '))
    p = (a + b + c) / 2
    area = (p * (p - a) * (p - b) * (p - c)) ** (1 / 2)
    return area


def calculate_area(type_of_fig: str) -> str:
    """ 
    Calculates area of one of the following figures:
    square, triangle, circle.   
    """
    
    area = 0
    type_of_fig = type_of_fig.lower()
    if type_of_fig == 'квадрат':
        area = calc_square_area()
    elif type_of_fig == 'круг':
        area = calc_circle_area()
    elif type_of_fig == 'треугольник':
        area = calc_triangle_area()
    else:
        return 'Такой фигуры нет. Попробуйте один из трёх вариантов: квадрат, круг, треугольник.'
    return f'Площадь {type_of_fig}а равна {area:.2f}'
    
    
while True:
    print('Программа вычисляет площадь квадрата, круга или треугольника.')
    print('Если хотите завершить, введите символ "q".')
    input_data = input('Введите один из доступных типов фигур: ')
    if input_data == 'q':
        break
    else:
        answ = calculate_area(input_data)
        print(answ)
        
        
# Test circle.
radius = 10
pi = 3.1416
area_1 = pi * (radius ** 2)
assert area_1 == 314.16

# Test triangle.
a = 2
b = 2
c = 3
p = (a + b + c) / 2
area_2 = (p * (p - a) * (p - b) * (p - c)) ** (1 / 2)
assert area_2 == 1.98

# Test square.
side_len = 10
area_3 = side_len ** 2
assert area_3 == 100
```
